```python
import pandas as pd
import numpy as np
import random
```

```python
col=['name','sex','births','year']
names=pd.read_excel('data/data1880.xlsx',header=None,names=col)
names.head(10)
```

```python
names['year']=np.random.randint(1880,2010)
names
```

```python
#聚合年份和性别数据
total_births=names.pivot_table('births',index='year',columns='sex',aggfunc=sum)
total_births
```

```python
total_births.plot(title='Total births by sex and year',)
```
